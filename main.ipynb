# Análisis Exploratorio con Visualizaciones Avanzadas

En esta sección analizaremos el comportamiento de las variables mediante diferentes tipos de gráficas que nos permitirán identificar patrones, distribuciones y relaciones entre las variables y el estado de diabetes.

## 1. Gráficas de Caja y Bigotes (Box Plots)

Los box plots nos permiten visualizar la distribución de variables continuas y identificar valores atípicos, mediana, cuartiles y la dispersión de los datos según el estado de diabetes.

```python
# Seleccionar variables numéricas clave para análisis
numerical_vars = ['b_m_i', 'ment_hlth', 'phys_hlth', 'age']

# Crear box plots para variables numéricas vs diabetes
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
axes = axes.ravel()

for i, var in enumerate(numerical_vars):
    sns.boxplot(data=df, x='diabetes_binary', y=var, ax=axes[i])
    axes[i].set_title(f'Distribución de {var.upper()} por Estado de Diabetes')
    axes[i].set_xlabel('Diabetes (0=No, 1=Sí)')
    axes[i].set_ylabel(var.replace('_', ' ').title())

plt.tight_layout()
plt.show()

# Estadísticas descriptivas por grupo
print("Estadísticas descriptivas por estado de diabetes:")
print("="*60)
for var in numerical_vars:
    print(f"\n{var.upper()}:")
    stats = df.groupby('diabetes_binary')[var].describe()
    print(stats)
```

### Análisis de los Box Plots:

**Observaciones importantes:**

1. **BMI (Índice de Masa Corporal)**: Se observa una clara diferencia entre los grupos. Las personas con diabetes tienden a tener un BMI más alto, con una mediana superior y mayor presencia de valores atípicos en el rango superior.

2. **Salud Mental**: La distribución muestra que las personas con diabetes reportan ligeramente más días de mala salud mental, aunque la diferencia no es tan marcada como en el BMI.

3. **Salud Física**: Existe una diferencia notable, donde las personas con diabetes reportan significativamente más días de mala salud física, indicando una correlación entre diabetes y problemas físicos.

4. **Edad**: Las personas con diabetes tienden a ser de mayor edad, con una mediana más alta y una distribución que se inclina hacia edades superiores.

Estos patrones sugieren que el BMI y la edad son factores de riesgo importantes para la diabetes, mientras que la condición también se asocia con una peor percepción de la salud física.

## 2. Gráficas de Violín (Violin Plots)

Los violin plots combinan la información de los box plots con la densidad de distribución, mostrándonos la forma de la distribución de los datos y permitiendo identificar si hay múltiples modas o distribuciones asimétricas.

```python
# Crear violin plots para las mismas variables
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
axes = axes.ravel()

for i, var in enumerate(numerical_vars):
    sns.violinplot(data=df, x='diabetes_binary', y=var, ax=axes[i])
    axes[i].set_title(f'Distribución (Densidad) de {var.upper()} por Estado de Diabetes')
    axes[i].set_xlabel('Diabetes (0=No, 1=Sí)')
    axes[i].set_ylabel(var.replace('_', ' ').title())

plt.tight_layout()
plt.show()
```

### Análisis de los Violin Plots:

**Patrones de distribución identificados:**

1. **BMI**: 
   - Sin diabetes: Distribución más concentrada en valores normales (18-25)
   - Con diabetes: Distribución más amplia con mayor densidad en valores altos (>25), indicando prevalencia de sobrepeso/obesidad

2. **Salud Mental**:
   - Ambos grupos muestran alta concentración en 0 días (buena salud mental)
   - Personas con diabetes muestran una cola más larga hacia valores altos, indicando más días de mala salud mental

3. **Salud Física**:
   - Sin diabetes: Fuerte concentración en 0 días de mala salud física
   - Con diabetes: Distribución más dispersa con mayor densidad en valores altos, confirmando mayor impacto físico

4. **Edad**:
   - Sin diabetes: Distribución más uniforme entre grupos de edad jóvenes y medios
   - Con diabetes: Concentración marcada en edades mayores, confirmando la edad como factor de riesgo

Los violin plots revelan que las diferencias no solo están en las medianas, sino en las formas de distribución completas.

## 3. Histogramas 

Los histogramas nos permiten visualizar la distribución de frecuencias de las variables y comparar cómo se distribuyen entre personas con y sin diabetes.

```python
# Crear histogramas superpuestos para comparar distribuciones
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
axes = axes.ravel()

colors = ['skyblue', 'salmon']
labels = ['Sin Diabetes', 'Con Diabetes']

for i, var in enumerate(numerical_vars):
    for diabetes_status in [0, 1]:
        data = df[df['diabetes_binary'] == diabetes_status][var]
        axes[i].hist(data, alpha=0.7, bins=30, color=colors[diabetes_status],
                    label=labels[diabetes_status], density=True)

    axes[i].set_title(f'Distribución de {var.upper()}')
    axes[i].set_xlabel(var.replace('_', ' ').title())
    axes[i].set_ylabel('Densidad')
    axes[i].legend()
    axes[i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Histogramas individuales más detallados para variables categóricas
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# Variables categóricas de interés
categorical_vars = ['high_b_p', 'high_chol', 'gen_hlth', 'smoker', 'stroke', 'heart_diseaseor_attack']

for i, var in enumerate(categorical_vars):
    if i < 6:  # Solo 6 subplots disponibles
        row = i // 3
        col = i % 3

        # Crear tabla de contingencia
        contingency = pd.crosstab(df[var], df['diabetes_binary'], normalize='columns')

        # Crear gráfico de barras
        contingency.plot(kind='bar', ax=axes[row, col], color=['skyblue', 'salmon'])
        axes[row, col].set_title(f'Distribución de {var.replace("_", " ").title()}')
        axes[row, col].set_xlabel(var.replace('_', ' ').title())
        axes[row, col].set_ylabel('Proporción')
        axes[row, col].legend(['Sin Diabetes', 'Con Diabetes'])
        axes[row, col].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

### Análisis de los Histogramas:

**Variables Numéricas:**

1. **BMI**: Clara separación de distribuciones. Sin diabetes muestra una distribución más normal centrada en valores saludables, mientras que con diabetes se desplaza hacia valores más altos con mayor dispersión.

2. **Salud Mental**: Ambos grupos concentrados en 0 días, pero diabetes muestra mayor variabilidad y frecuencia en días altos de mala salud mental.

3. **Salud Física**: Diferencia muy marcada. Sin diabetes altamente concentrado en 0 días, con diabetes muestra distribución más amplia hacia valores altos.

4. **Edad**: Sin diabetes relativamente uniforme en edades jóvenes-medias, con diabetes concentrado en edades mayores.

**Variables Categóricas:**

- **Presión Alta y Colesterol Alto**: Mayor prevalencia en personas con diabetes
- **Salud General**: Personas con diabetes reportan peor estado de salud general
- **Factores de Riesgo**: Mayor asociación de diabetes con enfermedades cardíacas y accidentes cerebrovasculares

Los histogramas confirman que la diabetes se asocia con un perfil de mayor riesgo en múltiples dimensiones de salud.

## 4. Resumen y Conclusiones del Análisis Visual

### Principales Hallazgos:

**Factores de Riesgo Identificados:**

1. **BMI Elevado**: Es el factor más claramente diferenciador. Las personas con diabetes muestran consistentemente valores más altos de BMI, indicando que el sobrepeso/obesidad es un factor de riesgo crítico.

2. **Edad Avanzada**: La diabetes es más prevalente en personas de mayor edad, con una distribución claramente sesgada hacia edades superiores.

3. **Comorbilidades**: Alta asociación con presión arterial alta, colesterol elevado, enfermedades cardíacas y accidentes cerebrovasculares.

4. **Impact en Calidad de Vida**: Las personas con diabetes reportan significativamente más días de mala salud física, y en menor medida, más problemas de salud mental.

### Implicaciones para la Prevención:

- **Control de Peso**: El BMI emerge como un predictor crucial y modificable
- **Monitoreo en Edad Avanzada**: Mayor vigilancia en poblaciones de edad mayor
- **Enfoque Integral**: La diabetes no es un problema aislado sino parte de un síndrome metabólico más amplio
- **Calidad de Vida**: Importancia del manejo integral que considere tanto aspectos físicos como mentales

### Limitaciones del Análisis:

- Los datos son transversales, no permiten establecer causalidad
- Posibles factores confundidores no considerados (genética, medicamentos, etc.)
- Auto-reporte puede introducir sesgos en variables como salud mental y física

Este análisis visual proporciona una base sólida para entender los patrones asociados con la diabetes y puede guiar estrategias de prevención y tratamiento.
