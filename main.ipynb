# Diabetes analysis project


This notebook will be use to display and analyze information about the diabetes and all the variables related with.

## Path Dir

```python
import os
import sys

sep = os.path.sep
dir_actual = os.path.abspath('')
PATH = sep.join(dir_actual.split(sep)[:-1])
DIR_DATA = PATH + '{0}diabetes_analysis_project{0}data{0}'.format(os.sep)
sys.path.append(PATH) if PATH not in list(sys.path) else None
DIR_DATA
```

## Importing libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
filename = DIR_DATA + 'diabetes_binary_health_indicators_BRFSS2021.csv'
```

## Load dataset

```python
df = pd.read_csv(filename, sep=',')
df
```

## Explore dataset structure

```python
print("Dataset dimensions (rows, columns):", df.shape)
print("\nFirst rows of the dataset:")
print(df.head())

print("\nData types of each column:")
print(df.dtypes)
```

## Show original column names

```python
print("\nOriginal column names:")
print(df.columns.tolist())
```

## Check for missing values and duplicates


```python
print("\nMissing values per column:")
print(df.isnull().sum())
```

>As we can see, there are no missing values, which represents an ideal situation.

```python
print("\nNumber of duplicated rows:", df.duplicated().sum())
```

>In this case, duplicates do not suppose a problem because they only indicate that there are people with the same or similar medical conditions.

## Standardize column names (optional but useful for analysis)

```python
# Convert to lowercase, insert underscores before capital letters, and remove spaces
df.columns = (
    df.columns
    .str.replace(r'(?<!^)(?=[A-Z])', '_', regex=True)  # add underscore before capital letters
    .str.replace(" ", "_")                             # replace spaces with underscores
    .str.lower()                                       # convert everything to lowercase
)

print("\nStandardized column names (snake_case):")
print(df.columns.tolist())

```

```python
df.head()
```

> Now we can see the dataset headers in snake_case

## Droping the columns we don't want

```python
columns_to_drop = ['no_docbc_cost', 'income', 'education']
df = df.drop(columns=columns_to_drop, errors='ignore')

print("\nColumns dropped:", columns_to_drop)
print("Remaining columns:")
print(df.columns.tolist())
```

```python
df
```

## Analyze target variable (diabetes_binary)

```python
# Count values of diabetes_binary
counts = df['diabetes_binary'].value_counts()

# Map labels (0 = No diabetes, 1 = Diabetes)
labels = ['No Diabetes (0)', 'Diabetes (1)']

# Create pie chart
plt.figure(figsize=(6,6))
plt.pie(
    counts,
    labels=labels,
    autopct='%1.1f%%',    # show percentages with 1 decimal
    colors=['skyblue', 'salmon'],
    startangle=90,
    wedgeprops={'edgecolor': 'white'}
)
plt.title("Distribution of Diabetes in the Dataset")
plt.show()

# Print absolute counts
print("Absolute counts of diabetes_binary:")
print(counts)
```
